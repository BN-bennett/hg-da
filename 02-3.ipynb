## 전체 책의 쪽수 구하기

```python
def get_page_cnt(isbn):
    # 교보문고 도서 상세 페이지 URL
    url = 'https://www.kyobobook.co.kr/product/detailViewKor.laf?barcode={}'
    # URL에 ISBN을 넣어 HTML 가져옵니다.
    r = requests.get(url.format(isbn))
    soup = BeautifulSoup(r.text, 'html.parser')   # HTML 파싱
    # 상품 상세정보 테이블을 선택합니다.
    prd_info = soup.find('table', attrs={'summary':'이 상품의 상세정보'})
    # 테이블에 있는 td 태그를 가져옵니다.
    prd_td_list = prd_info.find_all('td')
    return prd_td_list[1].get_text()  # 두 번째 td 태그에 있는 텍스트를 반환합니다.
```

```python
get_page_cnt(9791190090018)
```

```python
top10_books = books.head(10)
```

```python
page_count = top10_books.apply(lambda row: get_page_cnt(row['isbn13']), axis=1)
```

```python
def get_page_cnt2(row):
    isbn = row['isbn13']
    return get_page_cnt(isbn)
```

```python
top10_books.apply(get_page_cnt2, axis=1)
```

```python
print(page_count)
```

```python
page_count.name = 'page_count'

print(page_count)
```

```python
top10_with_page_count = pd.merge(top10_books, page_count, 
                                 left_index=True, right_index=True)

top10_with_page_count
```
